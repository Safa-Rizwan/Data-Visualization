# **Data Visualization Using Matplotlib**

# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Loading Dataset

```python
data = pd.read_csv('/content/D3_Heart_Dataset.csv')
data
```

# Line Graphs

```python
x = data["Age"]
y = data["Cholesterol"]
plt.plot(x,y)
plt.title("Age vs Cholesterol")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.show()
```

# Line Graph with subplot()

```python
x = data["Age"]
y1 = data["Cholesterol"]
y2 = data["MaxHR"]
y3 = data["RestingBP"]

# plot1
plt.subplot(1,3,1)
plt.plot(x,y1)
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.title("Age vs Cholesterol")

# plot2
plt.subplot(1,3,2)
plt.plot(x,y2)
plt.xlabel("Age")
plt.ylabel("MaxHR")
plt.title("Age vs MaxHR")

# plot3
plt.subplot(1,3,3)
plt.plot(x,y3)
plt.xlabel("Age")
plt.ylabel("RestingBP")
plt.title("Age vs RestingBP")

plt.show()
```

# Bar Graphs

```python
x = data["Gender"]
y = data["Age"]

# Plot
plt.bar(x,y)
plt.xlabel("Gender")
plt.ylabel("Age")
plt.title("Gender Vs Age")
plt.show()
```

# Male and Female Counts

```python
# Male
male = data["Gender"] == "M"
male_data = data[male].value_counts()
male_count = male_data.size

# Female
female = data["Gender"] == "F"
female_data = data[female].value_counts()
female_count = female_data.size

x = ["Male", "Female"]
y = [male_count , female_count]

plt.bar(x,y)
plt.show()
```
